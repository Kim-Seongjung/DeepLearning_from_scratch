### 이둘의 차이점은 어떻게 계산해야 할까요?
많은 방법이 있지만 우리는 아래와 같은 방법을 사용합니다.

MSE (Mean Squared Error)

```python

def mean_squared_error(y,t):
    return 0.5*np.sum((y-t)**2)
```

```python
t=np.array([0,1])
p=np.array([0.3,0.7])
print np.sum(t)
print np.sum(p)
mean_squared_error(t,p)
```

```python
def cross_entropy(y,t):
    delta = 1e-7
    return -np.sum(t*np.log(y+delta))

```

```python
x=np.arange(-2,2,0.01)
y=np.log(x)
y_=-np.log(x)
plt.ylim(-3,3)
plt.plot(x,y)
plt.plot(x,y_)
plt.title('log(x) Vs -log(x)')
```

```python
x=np.arange(-2,2,0.01)
y=np.log(x)
y_=-np.log(x)
plt.ylim(0,6)
plt.xlim(0,1)
plt.plot(x,y_)
plt.title('-log(x)')
```

```python
cross_entropy(1,0.6)
```
